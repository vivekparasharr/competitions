```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset with real seasonality
df = sns.load_dataset("flights")

print(df.head())  # always inspect

# Aggregate seasonality
seasonality = df.groupby("month")["passengers"].mean().reset_index()

# Plot
plt.figure(figsize=(10,5))
plt.plot(seasonality["month"], seasonality["passengers"], marker="o")

plt.title("Seasonality Pattern (Air Passengers by Month)")
plt.xlabel("Month")
plt.ylabel("Average Passengers")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```
