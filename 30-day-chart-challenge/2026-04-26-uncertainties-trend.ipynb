```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)

# Time
years = np.arange(2000, 2025)

# True trend (hidden reality)
true_trend = 0.03 * (years - 2000)

# Noisy observations (space sensor style)
noise = np.random.normal(0, 0.3, len(years))
observed = true_trend + noise

# Smooth trend (rolling mean = simple estimator)
trend = pd.Series(observed).rolling(window=5, center=True).mean()

# Uncertainty (approx = rolling std)
uncertainty = pd.Series(observed).rolling(window=5, center=True).std()

# Bounds
upper = trend + uncertainty
lower = trend - uncertainty

# Plot
plt.figure(figsize=(12,6))

plt.plot(years, observed, alpha=0.4, label="Observed (Noisy)")
plt.plot(years, trend, linewidth=3, label="Estimated Trend")

plt.fill_between(years, lower, upper, alpha=0.3, label="Uncertainty Band")

plt.title("Trend with Uncertainty (Space-Time Series)")
plt.xlabel("Year")
plt.ylabel("Signal")
plt.legend()
plt.grid(True)

plt.show()
```
