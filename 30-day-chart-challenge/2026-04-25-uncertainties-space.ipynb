```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# Simulated time (years)
years = np.arange(2000, 2025)

# Simulated satellite measurement (trend + noise)
true_signal = 0.02 * (years - 2000)  # warming trend

noise = np.random.normal(0, 0.2, len(years))

observed = true_signal + noise

# Uncertainty (instrument + atmospheric error)
uncertainty = np.linspace(0.3, 0.1, len(years))  # improving sensors over time

# Upper and lower bounds
upper = observed + uncertainty
lower = observed - uncertainty

# Plot
plt.figure(figsize=(12,6))

plt.plot(years, observed, label="Satellite Measurement")
plt.fill_between(years, lower, upper, alpha=0.3, label="Uncertainty Range")

plt.title("Space Time Series with Measurement Uncertainty")
plt.xlabel("Year")
plt.ylabel("Signal (arbitrary units)")
plt.legend()
plt.grid(True)

plt.show()
```
