```python
import networkx as nx
import matplotlib.pyplot as plt

# -----------------------------
# Central media organization
# -----------------------------
center = "South China Morning Post"

# -----------------------------
# Relationship structure:
# SCMP → regions + topics it covers
# -----------------------------
edges = [
    (center, "China"),
    (center, "Hong Kong"),
    (center, "Taiwan"),
    (center, "United States"),
    (center, "Global Trade"),
    (center, "Geopolitics"),
    (center, "Technology"),
    (center, "Economy"),
    (center, "South China Sea"),
    (center, "Artificial Intelligence"),
    (center, "Elections"),
    (center, "Climate Policy")
]

# Build graph
G = nx.Graph()
G.add_edges_from(edges)

# -----------------------------
# Layout (radial feel)
# -----------------------------
plt.figure(figsize=(10, 7))
pos = nx.spring_layout(G, seed=42)

# -----------------------------
# Draw nodes
# -----------------------------
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color="lightblue")

# Highlight central node
nx.draw_networkx_nodes(
    G, pos,
    nodelist=[center],
    node_size=3000,
    node_color="orange"
)

# -----------------------------
# Draw edges
# -----------------------------
nx.draw_networkx_edges(G, pos, width=1.5, alpha=0.6)

# Labels
nx.draw_networkx_labels(G, pos, font_size=9)

plt.title("Relationships → Media Coverage Network (SCMP)")
plt.axis("off")
plt.show()
```
