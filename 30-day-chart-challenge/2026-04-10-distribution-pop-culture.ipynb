```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Sample Pop Culture Dataset
# -----------------------------
data = {
    "category": [
        "Movies", "Music", "TV Shows", "Video Games",
        "Memes", "Books", "Podcasts", "Fashion"
    ],
    "share": [22, 18, 15, 14, 10, 8, 7, 6]
}

df = pd.DataFrame(data)

# -----------------------------
# Figure Setup
# -----------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# -----------------------------
# 1. Donut Chart (Pop Culture Mix)
# -----------------------------
axes[0].pie(
    df["share"],
    labels=df["category"],
    autopct="%1.1f%%",
    startangle=140,
    wedgeprops={"width": 0.4}  # donut style
)

axes[0].set_title("Pop Culture Distribution (Donut Chart)")

# -----------------------------
# 2. Bar Chart (Comparison View)
# -----------------------------
axes[1].bar(df["category"], df["share"])
axes[1].set_title("Pop Culture Distribution (Bar Chart)")
axes[1].set_ylabel("Share (%)")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
