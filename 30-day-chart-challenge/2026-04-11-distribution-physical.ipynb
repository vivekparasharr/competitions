```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Physical world distribution dataset
# -----------------------------
data = {
    "land_type": [
        "Forests",
        "Water Bodies",
        "Urban Areas",
        "Agriculture",
        "Mountains",
        "Deserts",
        "Grasslands"
    ],
    "coverage": [31, 26, 12, 18, 7, 4, 2]
}

df = pd.DataFrame(data)

# Sort for better visualization
df = df.sort_values("coverage", ascending=False)

# -----------------------------
# Figure setup
# -----------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# -----------------------------
# 1. Donut Chart (Physical Distribution Mix)
# -----------------------------
axes[0].pie(
    df["coverage"],
    labels=df["land_type"],
    autopct="%1.1f%%",
    startangle=140,
    wedgeprops={"width": 0.4}
)

axes[0].set_title("Physical World Distribution (Donut Chart)")

# -----------------------------
# 2. Horizontal Bar Chart
# -----------------------------
axes[1].barh(df["land_type"], df["coverage"])
axes[1].set_title("Physical Distribution (Bar Chart)")
axes[1].set_xlabel("Coverage (%)")

plt.tight_layout()
plt.show()
```
