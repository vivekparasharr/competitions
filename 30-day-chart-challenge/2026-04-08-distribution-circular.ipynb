```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Olympic-style dataset
data = pd.DataFrame({
    "Country": ["USA", "China", "Japan", "UK", "ROC"],
    "Gold": [39, 38, 27, 22, 20],
    "Silver": [41, 32, 14, 21, 28],
    "Bronze": [33, 18, 17, 22, 23]
})

# Compute total medals
data["Total"] = data[["Gold", "Silver", "Bronze"]].sum(axis=1)

# Number of categories
N = len(data)

# Angle positions (circular layout)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False)

# Close the circle
angles = np.append(angles, angles[0])
totals = np.append(data["Total"].values, data["Total"].values[0])
labels = np.append(data["Country"], data["Country"][0])

# Create polar plot
fig = plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Plot bars
ax.bar(angles[:-1], data["Total"], width=0.5, alpha=0.7)

# Add labels
for i, label in enumerate(data["Country"]):
    ax.text(angles[i], data["Total"][i] + 5, label, ha='center', va='center')

# Styling
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

ax.set_title("Circular Distribution: Olympic Total Medals by Country", pad=20)

# Remove radial ticks for cleaner FlowingData style
ax.set_yticklabels([])

plt.tight_layout()
plt.show()
```
