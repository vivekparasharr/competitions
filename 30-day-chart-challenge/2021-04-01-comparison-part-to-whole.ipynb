```python
import plotly.graph_objects as go

# Data
labels = ['All other movies', 'Indian movies']
values = [6357, 923]
colors = ['#636EFA', '#EF553B']  # Custom colors

# Helper function for annotations
def create_annotation(text, y, size=14, color='grey', family='Arial'):
    return dict(
        xref='paper', yref='paper',
        x=0.5, y=y,
        xanchor='center', yanchor='top',
        font=dict(family=family, size=size, color=color),
        showarrow=False,
        text=text
    )

# Pie chart
fig = go.Figure(
    data=[go.Pie(
        labels=labels,
        values=values,
        pull=[0, 0.2],
        rotation=90,
        marker=dict(colors=colors),
        hole=0.0,           # 0 for full pie, >0 for donut
        hoverinfo='label+percent+value'
    )]
)

# Layout with title, subtitle, footer
fig.update_layout(
    annotations=[
        create_annotation("Share of Indian movies on Netflix", y=1.22, size=24),
        create_annotation("Netflix Movies and TV Shows Dataset", y=1.07, size=14),
        create_annotation("#30DayChartChallenge - part-to-whole - 2021/04/01", y=-0.0, size=12),
        create_annotation("Dataset from Kaggle: https://www.kaggle.com/shivamb/netflix-shows", y=-0.06, size=12),
        create_annotation("twitter.com/vivekparasharr | github.com/vivekparasharr", y=-0.15, size=12)
    ],
    margin=dict(t=150, b=100)  # Add space for annotations
)

# Show figure
fig.show()
```
