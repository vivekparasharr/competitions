```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Real-world dataset (Gapminder-style GDP per capita data)
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminder_with_codes.csv"
df = pd.read_csv(url)

# Use latest available year data proxy (filter for 2007, last complete in dataset)
df_2007 = df[df["year"] == 2007].copy()

# Wealth proxy
wealth = df_2007["gdpPercap"]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(wealth, bins=30, color="steelblue", edgecolor="white")

ax.set_title("Global Wealth Distribution (GDP per capita, 2007)")
ax.set_xlabel("GDP per capita (USD)")
ax.set_ylabel("Number of Countries")

# FlowingData-style clean look
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```
