```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

edible_plants = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2026/2026-02-03/edible_plants.csv')
edible_plants.tail(2)   
```

```python
# Distribution by sunlight
df = edible_plants["sunlight"].value_counts().reset_index()
df.columns = ["sunlight", "count"]

# Normalize to 100 icons
total_icons = 100
df["icons"] = (df["count"] / df["count"].sum() * total_icons).round().astype(int)

# Plot
fig, ax = plt.subplots(figsize=(12, 5))

x_offset = 0
cols = 10

for _, row in df.iterrows():
    n = row["icons"]
    rows = int(np.ceil(n / cols))

    xs = np.tile(np.arange(cols), rows)[:n] + x_offset
    ys = -np.repeat(np.arange(rows), cols)[:n]

    # draw emoji as text instead of marker
    for x, y in zip(xs, ys):
        ax.text(x, y, "🌱", fontsize=14, ha="center", va="center")

    # category label
    ax.text(
        x_offset + cols / 2,
        1,
        f"{row['sunlight']} ({row['count']})",
        ha="center",
        fontsize=11
    )

    x_offset += cols + 3

ax.set_title("Edible Plants by Sunlight Requirement (Pictogram)", fontsize=14)
ax.axis("off")

plt.tight_layout()
plt.show()
```
