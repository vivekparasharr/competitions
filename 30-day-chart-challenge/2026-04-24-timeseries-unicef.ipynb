```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Built-in stable dataset (NO URL issues)
df = sns.load_dataset("healthexp")

print(df.head())

# Pick one country (you can change this)
country = "Canada"
data = df[df["Country"] == country]

# Sort by year
data = data.sort_values("Year")

# Plot
plt.figure(figsize=(12,6))
plt.plot(data["Year"], data["Life_Expectancy"], marker="o")

plt.title(f"UNICEF-Style Indicator: Life Expectancy in {country}")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.grid(True)

plt.show()
```
