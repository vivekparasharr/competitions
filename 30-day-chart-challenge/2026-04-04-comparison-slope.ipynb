```python
import pandas as pd
import matplotlib.pyplot as plt

# Data: Gold medals comparison (realistic values)
data = pd.DataFrame({
    "Country": ["USA", "China", "Japan", "UK", "Russia"],
    "2016": [46, 26, 12, 27, 19],
    "2020": [39, 38, 27, 22, 20]
})

# Sort by 2020 for cleaner look
data = data.sort_values(by="2020", ascending=True).reset_index(drop=True)

# Plot
fig, ax = plt.subplots(figsize=(8, 6))

for i in range(len(data)):
    x = [0, 1]
    y = [data.loc[i, "2016"], data.loc[i, "2020"]]
    
    ax.plot(x, y, marker="o")
    
    # Labels
    ax.text(-0.05, y[0], f"{data.loc[i, 'Country']} ({y[0]})", ha='right', va='center')
    ax.text(1.05, y[1], f"{y[1]}", ha='left', va='center')

# Formatting
ax.set_xticks([0, 1])
ax.set_xticklabels(["2016", "2020"])
ax.set_title("Olympic Gold Medals Change (2016 → 2020)")

# Clean look
for spine in ["top", "right", "left"]:
    ax.spines[spine].set_visible(False)

ax.tick_params(axis='y', left=False)

plt.tight_layout()
plt.show()
```
