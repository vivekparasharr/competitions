```python
import pandas as pd
import matplotlib.pyplot as plt

# Experimental dataset (Olympics comparison)
data = pd.DataFrame({
    "Country": ["USA", "China", "Japan", "UK", "Russia"],
    "2016": [46, 26, 12, 27, 19],
    "2020": [39, 38, 27, 22, 20]
})

# Sort by improvement (2020 - 2016)
data["diff"] = data["2020"] - data["2016"]
data = data.sort_values("diff")

# Plot
fig, ax = plt.subplots(figsize=(9, 6))

for i in range(len(data)):
    y = i
    x1 = data.loc[i, "2016"]
    x2 = data.loc[i, "2020"]
    
    # Line connecting two experiments
    ax.plot([x1, x2], [y, y], color="gray", linewidth=2)
    
    # Points (Experiment A & B)
    ax.scatter(x1, y, color="blue", s=80, label="2016" if i == 0 else "")
    ax.scatter(x2, y, color="green", s=80, label="2020" if i == 0 else "")
    
    # Labels
    ax.text(x1 - 1, y, f"{x1}", va="center", ha="right")
    ax.text(x2 + 1, y, f"{x2}", va="center", ha="left")
    ax.text(max(x1, x2) + 6, y, data.loc[i, "Country"], va="center")

# Formatting
ax.set_yticks(range(len(data)))
ax.set_yticklabels([""] * len(data))
ax.set_xlabel("Gold Medals")

ax.set_title("Experimental Comparison: Olympic Performance (2016 vs 2020)")

ax.legend()

# Clean look
for spine in ["top", "right", "left"]:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```
