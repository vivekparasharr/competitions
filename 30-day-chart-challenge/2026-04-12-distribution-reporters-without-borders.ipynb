```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Sample RSF-style dataset
# (Press Freedom Index scores)
# Higher = worse press freedom
# -----------------------------
data = {
    "country": [
        "Norway", "Sweden", "Finland", "Netherlands",
        "Germany", "Canada", "UK", "USA",
        "France", "Japan", "India", "Brazil",
        "Mexico", "Turkey", "Russia", "China"
    ],
    "rsf_score": [
        8, 9, 9, 12,
        15, 16, 18, 22,
        20, 24, 36, 31,
        43, 48, 55, 75
    ]
}

df = pd.DataFrame(data)

# Sort by score
df = df.sort_values("rsf_score", ascending=True)

# -----------------------------
# Plot setup
# -----------------------------
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# -----------------------------
# 1. Bar chart (country comparison)
# -----------------------------
axes[0].bar(df["country"], df["rsf_score"])
axes[0].set_title("RSF Press Freedom Index (Lower = Better)")
axes[0].set_ylabel("Score")
axes[0].tick_params(axis='x', rotation=60)

# -----------------------------
# 2. Distribution histogram
# -----------------------------
axes[1].hist(df["rsf_score"], bins=6, edgecolor="black")
axes[1].set_title("Distribution of Press Freedom Scores")
axes[1].set_xlabel("RSF Score")
axes[1].set_ylabel("Number of Countries")

plt.tight_layout()
plt.show()
```
