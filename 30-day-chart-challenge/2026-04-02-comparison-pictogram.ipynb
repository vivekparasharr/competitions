```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Segoe UI Emoji"   # Windows
# plt.rcParams["font.family"] = "Apple Color Emoji"  # Mac
# plt.rcParams["font.family"] = "Noto Color Emoji"   # Linux

# Data
data = pd.DataFrame({
    "Country": ["USA", "China", "Japan", "UK", "ROC"],
    "Gold": [39, 38, 27, 22, 20]
})

scale = 2
data["icons"] = (data["Gold"] / scale).round().astype(int)

# Plot
fig, ax = plt.subplots(figsize=(12, 6))

for i in range(len(data)):
    for j in range(data.loc[i, "icons"]):
        ax.text(j, i, "🥇", fontsize=18)

# force proper layout
ax.set_xlim(0, data["icons"].max() + 2)
ax.set_ylim(-1, len(data))

# Labels
ax.set_yticks(range(len(data)))
ax.set_yticklabels(data["Country"])
ax.set_xticks([])

ax.set_title("Olympic Gold Medals (Pictogram)", fontsize=14)

# Clean look
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```
