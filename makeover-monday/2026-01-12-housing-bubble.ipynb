### The Biggest Housing Bubble Risks Globally

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('https://query.data.world/s/g2vubhd5worhcabmf5f56lk4b4cilg?dws=00000')
df.tail(2)
```

```python
df = df.sort_values("Bubble Risk Score", ascending=True)

sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Horizontal bar chart
sns.barplot(
    data=df,
    x="Bubble Risk Score",
    y="City",
    hue="Bubble Risk Category",
    dodge=False,
    ax=axes[0, 0]
)
axes[0, 0].set_title("Bubble Risk Score by City")

# 2. Scatter plot
sns.scatterplot(
    data=df,
    x="Rank",
    y="Bubble Risk Score",
    hue="Bubble Risk Category",
    ax=axes[0, 1]
)
axes[0, 1].set_title("Rank vs Bubble Risk Score")

# 3. Box plot
sns.boxplot(
    data=df,
    x="Bubble Risk Category",
    y="Bubble Risk Score",
    ax=axes[1, 0]
)
axes[1, 0].set_title("Score Distribution by Category")

# 4. Count plot
sns.countplot(
    data=df,
    x="Bubble Risk Category",
    ax=axes[1, 1]
)
axes[1, 1].set_title("Number of Cities per Category")

# Master title
fig.suptitle(
    "Global Housing Bubble Risk Analysis",
    fontsize=16,
    fontweight="bold"
)

# Adjust layout so title doesn’t overlap
plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```
